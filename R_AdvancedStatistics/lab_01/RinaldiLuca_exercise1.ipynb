### Exercise 3
***

#### 3.1)

```r
coal_t <- read_csv('DATA/coal-production-by-country.csv')
coal_t <- as.tibble(coal_t)
```

#### 3.2)

```r
unique_country <- unique(coal_t$Entity)
cat('Number of country:',length(unique_country))
```

```r
y <- integer(length(unique_country))
for (i in 1:length(unique_country)) 
    y[i] <- length(coal_t$Entity[coal_t$Entity==unique_country[i]])

options(repr.plot.width=10, repr.plot.height=40)
par(mar=c(5,12,4,1))
barplot(y, names.arg = unique_country, horiz = T,las=1)
```

#### 3.3)

```r
new_coal_t <- filter(coal_t,Year>=1970) %>% group_by(Entity) %>% summarise(integ = sum(`Coal production (TWh)`)*24*365)
```

```r
cat('Top 5 country for coal production:')
top_country <- new_coal_t$Entity[order(new_coal_t$integ, decreasing = T)][c(4,7,11,13,14)]
top_country
```

#### 3.4)

```r
top5 <- filter(coal_t,Entity %in% top_country,Year>=1970)

options(repr.plot.width=15, repr.plot.height=6)
ggplot(top5, aes(x=Year, y=`Coal production (TWh)`, color=Entity)) + geom_point() + geom_line()
```

#### 3.5)

```r
world_t <- filter(coal_t,Year>=1970,Entity=='World') %>% group_by(Entity,Year) %>% summarise(cumul = cumsum(`Coal production (TWh)`))
ggplot(world_t, aes(x=Year, y=cumul)) + geom_point() + geom_line()
```

### Exercise 4

```r
fname <- 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations-by-manufacturer.csv'
covid_t <- read_csv(file = url(fname))
covid_t <- as_tibble(covid_t)
```

#### 4.1)

```r
covid_it <- filter(covid_t, location=='Italy')
```

#### 4.2)

```r
plt <- ggplot(covid_it, aes(x=date, y=total_vaccinations, col=vaccine)) + geom_point()
plt + labs(title='Vaccination in Italy', y='Total number of vaccinations', x='Date')
```

#### 4.3)

```r
covid_it_daily <- covid_it %>% group_by(date) %>% summarise(shots_daily = sum(total_vaccinations))
plt <- ggplot(covid_it_daily, aes(x=date, y=shots_daily)) + geom_point()
plt + labs(title='Vaccines shots per day',y='Total number of vaccinations', x='Date')
```

#### 4.4)

```r
#Germany
covid_de <- filter(covid_t, location=='Germany')
plt <- ggplot(covid_de, aes(x=date, y=total_vaccinations, col=vaccine)) + geom_point()
plt + labs(title='Vaccination in Germany', y='Total number of vaccinations', x='Date')
covid_de_daily <- covid_de %>% group_by(date) %>% summarise(shots_daily = sum(total_vaccinations))
plt2 <- ggplot(covid_de_daily, aes(x=date, y=shots_daily)) + geom_point()
plt2 + labs(title='Vaccines shots per day',y='Total number of vaccinations', x='Date')
```

```r
#US
covid_us <- filter(covid_t, location=='United States')
plt <- ggplot(covid_us, aes(x=date, y=total_vaccinations, col=vaccine)) + geom_point()
plt + labs(title='Vaccination in United States', y='Total number of vaccinations', x='Date')
covid_us_daily <- covid_us %>% group_by(date) %>% summarise(shots_daily = sum(total_vaccinations))
plt2 <- ggplot(covid_us_daily, aes(x=date, y=shots_daily)) + geom_point()
plt2 + labs(title='Vaccines shots per day',y='Total number of vaccinations', x='Date')
```
