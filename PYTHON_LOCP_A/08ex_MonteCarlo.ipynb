1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
np.random.seed(1234)

def prob(x):
    return 1 - 2**(-x/(3.052*60))

tau = 3.052*60
Tl  = np.zeros(2500)
Pb  = np.zeros(2500)
Tl[0] = 1000
Pb[0] = 0

for t in range(1,2500):
    p = 2**(-1/tau) # probability to not decay
    decay = np.where(np.random.random(int(Tl[t-1])) < 1-p)[0].size
    Tl[t] = Tl[t-1] - decay
    Pb[t] = Pb[t-1] + decay

time = np.arange(2500)
plt.figure(figsize=(12,4))
plt.scatter(time, Tl, label='Tl')
plt.scatter(time, Pb, label='Pb')
plt.plot(time, 1000*prob(time), 'r', label='1-2^(-t/tau)')
plt.xlabel('Time')
plt.legend(loc='best')
plt.show()
```

```python
def icdf(x):
    return -(3.052*60)*np.log2(1-x)


rnd = np.random.random(1000)
dec_t = icdf(rnd)
t = np.arange(int(dec_t.max())+5)
Tl = np.array([ np.where( dec_t > i)[0].size for i in t])

plt.figure(figsize=(12,4))
plt.scatter(t, Tl, label='Tl')
plt.plot(t, 1000-1000*prob(t), 'r', label='2^(-t/tau)')
plt.legend(loc='best')
plt.show()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
z     = 79
Energ = 7.7*1e6 # E/e
e     = 1.602176634*1e-19
eps0  = 8.8541878128*1e-12
a0    = 5.291777721092*1e-11

x = np.random.normal(loc=0, scale=a0/100, size=int(1e6))
y = np.random.normal(scale=a0/100, size=int(1e6))
b = np.sqrt(x*x + y*y)

theta = 2*np.arctan(z*e/(2*np.pi*eps0*Energ*b))
print('The fraction of the back-scattered particles is: %.6f' %(np.where(theta>np.pi/2)[0].size /1e6))
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def f(x):
    return np.sin(1/(x*(2-x)))**2

N = 100000
count = np.zeros(100)
for i in range(100):
    x = 2*np.random.random(N)
    y = np.random.random(N)
    count[i] = np.where(y < f(x))[0].size
I=2*count/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

```python
I = np.zeros(100)
for i in range(100):
    sample = 2*np.random.random(N)
    I[i] = 2*np.sum(f(sample))/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
# Slower method

def f_circle(x, y):
    return 1 if (x*x + y*y) <= 1 else 0

N = 100000
count = np.zeros(100)
for i in range(100):
    x = 2*np.random.random(N) -1
    y = 2*np.random.random(N) -1
    for j in range(N):
        count[i] += f_circle(x[j],y[j])
I = 4*count/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

```python
# Faster method

def f_circle(x, y):
    return x*x + y*y 

N = 100000
count = np.zeros(100)
for i in range(100):
    x = 2*np.random.random(N) -1
    y = 2*np.random.random(N) -1
    count[i] = np.where(f_circle(x,y) <= 1)[0].size
I = 4*count/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

```python
def f_ipershpere(x):
    return np.sum(x*x, axis=0)

dim = 10
N = 1000000
count = np.zeros(100)
for i in range(100):
    x = 2*np.random.random((dim,N)) -1
    count[i] = np.where(f_ipershpere(x) <= 1)[0].size
I= (2**dim)*count/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
def f(x):
    return x**(-1/2) / (np.exp(x)+1)

def cdf(x):
    return 2*x**(1/2)

def icdf(x):
    return (x/2)**2

N = 100000 
a = cdf(1)
I = np.zeros(100)
for i in range(100):
    sample = np.random.random(N)*a
    imp_sample = icdf(sample)
    I[i] = 2*np.sum(f(imp_sample) * np.sqrt(imp_sample))/N

print('Integral value, as mean over 100 computation: %.4f' %(I.mean()))
print('Error as std over 100 computation: %.4f' %(I.std()))
```
